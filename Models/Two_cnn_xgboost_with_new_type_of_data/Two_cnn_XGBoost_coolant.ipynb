```python
# Example for loop through grouped trips
for trip_id, trip_data in trip_groups:
    # Reset index for safe operations
    trip_data = trip_data.reset_index(drop=True)

    # Check if the trip length is greater than 1000
    if len(trip_data) > 2000:
        # Slice the trip data into chunks of 1000 rows
        for i in range(0, len(trip_data), 2000):
            # Get the chunk of 1000 rows (or less for the last chunk)
            chunk = trip_data.iloc[i:i+2000].reset_index(drop=True)

            # Subtract the first row's time from all rows to start from 0
            initial_time = chunk['time'].iloc[0]
            chunk['time'] = chunk['time'] - initial_time

            # Add a column with the initial coolant temperature (from the first row)
            initial_coolant_temp = chunk['Coolant_temperature'].iloc[0]
            chunk['initial_coolant_temp'] = initial_coolant_temp

            # Save the modified chunk to a separate CSV file
            chunk_file_name = f'{folder_path}/trip_{trip_id}_part_{i//2000 + 1}.csv'
            chunk.to_csv(chunk_file_name, index=False)

            print(f'Saved {chunk_file_name} with time reset and initial coolant temp')
    else:
        # If the trip has less than or equal to 1000 rows, process as a single file
        initial_time = trip_data['time'].iloc[0]
        trip_data['time'] = trip_data['time'] - initial_time

        # Add a column with the initial coolant temperature (from the first row)
        initial_coolant_temp = trip_data['Coolant_temperature'].iloc[0]
        trip_data['initial_coolant_temp'] = initial_coolant_temp

        # Save the modified trip data to a separate CSV file
        trip_file_name = f'{folder_path}/trip_{trip_id}.csv'
        trip_data.to_csv(trip_file_name, index=False)

        print(f'Saved {trip_file_name} with time reset and initial coolant temp')
```

```python
import os

# Example for loop through grouped trips
for trip_id, trip_data in trip_groups:
    # Reset index for safe operations
    trip_data = trip_data.reset_index(drop=True)

    # Initialize chunk index and start index
    chunk_idx = 1
    start_idx = 0

    # Loop through the trip data to check for non-monotonic behavior in time
    for i in range(1, len(trip_data)):
        # Check if time jumps (non-monotonic behavior)
        if trip_data['time'].iloc[i] < trip_data['time'].iloc[i - 1]:
            # If time jumps, create a new chunk from the last index to the current index
            chunk = trip_data.iloc[start_idx:i].reset_index(drop=True)

            # Only save the chunk if it has data
            if len(chunk) > 0:
                # Adjust time to start from 0 for the new chunk
                initial_time = chunk['time'].iloc[0]
                chunk['time'] = chunk['time'] - initial_time

                # Add a column with the initial coolant temperature (from the first row)
                initial_coolant_temp = chunk['Coolant_temperature'].iloc[0]
                chunk['initial_coolant_temp'] = initial_coolant_temp

                # Save the chunk to a separate CSV file
                chunk_file_name = f'{folder_path}/trip_{trip_id}_part_{chunk_idx}.csv'
                chunk.to_csv(chunk_file_name, index=False)
                print(f"Saved {chunk_file_name} with time reset and initial coolant temp")

            # Update the start index to the current point (i.e., after the jump)
            start_idx = i
            chunk_idx += 1

    # Handle the final chunk after the last detected jump (if any data remains)
    if start_idx < len(trip_data):
        chunk = trip_data.iloc[start_idx:].reset_index(drop=True)

        # Adjust time to start from 0 for the final chunk
        if len(chunk) > 0:
            initial_time = chunk['time'].iloc[0]
            chunk['time'] = chunk['time'] - initial_time

            # Add a column with the initial coolant temperature (from the first row)
            initial_coolant_temp = chunk['Coolant_temperature'].iloc[0]
            chunk['initial_coolant_temp'] = initial_coolant_temp

            # Save the final chunk
            chunk_file_name = f'{folder_path}/trip_{trip_id}_part_{chunk_idx}.csv'
            chunk.to_csv(chunk_file_name, index=False)
            print(f"Saved {chunk_file_name} with time reset and initial coolant temp")
```

# **Add change of speed and gear**

```python
import pandas as pd

# Parameters
time_jump_threshold = 1000  # Adjust as needed
max_chunk_size = 2000       # Maximum rows per chunk

for trip_id, trip_data in trip_groups:
    trip_data = trip_data.reset_index(drop=True)
    file_count = 1
    start_idx = 0
    prev_time = trip_data['time'].iloc[0]

    for idx in range(1, len(trip_data)):
        current_time = trip_data['time'].iloc[idx]

        # Check for time discontinuity or maximum chunk size
        if current_time < prev_time or (current_time - prev_time) > time_jump_threshold or (idx - start_idx) >= max_chunk_size:
            # Extract the chunk of data
            chunk = trip_data.iloc[start_idx:idx].reset_index(drop=True)

            # Reset time based on initial time
            initial_time = chunk['time'].iloc[0]
            chunk['time'] = chunk['time'] - initial_time

            # Add initial coolant temperature
            initial_coolant_temp = chunk['Coolant_temperature'].iloc[0]
            chunk['initial_coolant_temp'] = initial_coolant_temp

            # Add cumulative speed difference
            chunk['speed_diff'] = chunk['Vehicle_Speed'].diff().abs().fillna(0).cumsum()

            # Add cumulative gear change
            chunk['gear_change'] = chunk['Current_gear_shift_position_(Current_gear)'].diff().abs().fillna(0).cumsum()

            # Save the chunk to a CSV file
            chunk_file_name = f'{folder_path}/trip_{trip_id}_part_{file_count}.csv'
            chunk.to_csv(chunk_file_name, index=False)
            print(f'Saved {chunk_file_name} with time reset, initial coolant temp, speed diff, and gear changes')

            # Update counters for next chunk
            file_count += 1
            start_idx = idx
        prev_time = current_time

    # Save any remaining data
    if start_idx < len(trip_data):
        chunk = trip_data.iloc[start_idx:].reset_index(drop=True)

        # Reset time based on initial time
        initial_time = chunk['time'].iloc[0]
        chunk['time'] = chunk['time'] - initial_time

        # Add initial coolant temperature
        initial_coolant_temp = chunk['Coolant_temperature'].iloc[0]
        chunk['initial_coolant_temp'] = initial_coolant_temp

        # Add cumulative speed difference
        chunk['speed_diff'] = chunk['Vehicle_Speed'].diff().abs().fillna(0).cumsum()

        # Add cumulative gear change
        chunk['gear_change'] = chunk['Current_gear_shift_position_(Current_gear)'].diff().abs().fillna(0).cumsum()

        # Save the remaining chunk to a CSV file
        chunk_file_name = f'{folder_path}/trip_{trip_id}_part_{file_count}.csv'
        chunk.to_csv(chunk_file_name, index=False)
        print(f'Saved {chunk_file_name} with time reset, initial coolant temp, speed diff, and gear changes')
```

```python
import pandas as pd
import os

# Specify the folder where your CSV files are stored
folder_path = '/content/csv_data/'  # Replace with your folder path

# Get a list of all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Initialize an empty list to store DataFrames
dataframes = []

# Loop through each file and read it into a DataFrame, then append it to the list
for csv_file in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    df = pd.read_csv(file_path)
    if len(df) < 300:
            continue
    # Filter rows where time is non-negative and limit to 3000 rows
    df = df[df['time'] >= 0]  # Filter out rows where time is negative
    df = df[df['Coolant_temperature'] >= 0]
    # df = df.iloc[:1000]
    # If there are any valid rows, append to dataframes list
    if not df.empty:
        dataframes.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)

# Display the combined DataFrame (or save it if needed)
print(combined_df)

# Optionally, save the combined DataFrame to a CSV file
combined_df.to_csv(f'combined_trips_find_time_index_all.csv', index=False)
combined_df
```

```python
import numpy as np
import pandas as pd

# Function to add Gaussian noise to a column and clip negative values
def add_noise(data, noise_level=0.01):
    """
    Adds Gaussian noise to the data and clips negative values.
    :param data: pandas Series or numpy array
    :param noise_level: The standard deviation of the noise relative to the data range (percentage of data).
    :return: Data with added noise, ensuring no negative values
    """
    noise = np.random.normal(0, noise_level * data.std(), data.shape)
    noisy_data = data + noise
    return np.clip(noisy_data, a_min=0, a_max=None)  # Clip values to ensure no negatives (min=0)

# Apply noise to specific columns in the combined DataFrame
columns_to_augment = ['Vehicle_Speed', 'Current_gear_shift_position_(Current_gear)', 'time','initial_coolant_temp' ,'speed_diff' ,'gear_change']  # Add noise to these columns

# Create a copy of the original DataFrame to apply noise and create augmented data
augmented_df = combined_df.copy()

for column in columns_to_augment:
    augmented_df[column] = add_noise(augmented_df[column], noise_level=0.10)  # Adjust noise level if needed

# Concatenate the original and augmented data
augmented_combined_df = pd.concat([combined_df, augmented_df], ignore_index=True)

# Optionally, save the concatenated DataFrame to a new CSV file
augmented_combined_df.to_csv('combined_original_and_augmented_data.csv', index=False)

# Display the concatenated DataFrame
print(augmented_combined_df)
```

```python
import torch
import torch.nn as nn

# Helper function to calculate the output size after a convolution
def calculate_conv_output_size(input_size, kernel_size, stride=1, padding=0):
    return (input_size - kernel_size + 2 * padding) // stride + 1

class CoolantTempCNN(nn.Module):
    def __init__(self, input_sequence_length, kernel_size1=3, kernel_size2=15):
        super(CoolantTempCNN, self).__init__()
        # Compute padding for conv1 based on kernel_size1
        padding1 = (kernel_size1 - 1) // 2
        # First 1D convolutional layer with variable kernel size
        self.conv1 = nn.Conv1d(in_channels=4, out_channels=16, kernel_size=kernel_size1, stride=1, padding=padding1)
        self.bn1 = nn.BatchNorm1d(16)  # Batch normalization after conv1

        # Compute padding for conv2 based on kernel_size2
        padding2 = (kernel_size2 - 1) // 2
        # Second 1D convolutional layer with variable kernel size
        self.conv2 = nn.Conv1d(in_channels=16, out_channels=32, kernel_size=kernel_size2, stride=1, padding=padding2)
        self.bn2 = nn.BatchNorm1d(32)  # Batch normalization after conv2

        # Dropout layer to prevent overfitting
        self.dropout = nn.Dropout(p=0.2)

        # Calculate the sequence length after each convolution
        sequence_length_after_conv1 = calculate_conv_output_size(
            input_sequence_length, kernel_size=kernel_size1, padding=padding1)
        sequence_length_after_conv2 = calculate_conv_output_size(
            sequence_length_after_conv1, kernel_size=kernel_size2, padding=padding2)

        # Fully connected layers (adjust input size based on conv2 output)
        self.fc1 = nn.Linear(32 * sequence_length_after_conv2, 64)
        self.ln = nn.LayerNorm(64)  # Layer normalization before final layer
        self.fc2 = nn.Linear(64, 1)  # Regression output for predicting coolant temperature

    def forward(self, x):
        # Apply first convolutional layer + batch norm + relu
        x = torch.relu(self.bn1(self.conv1(x)))

        # Apply second convolutional layer + batch norm + relu
        x = torch.relu(self.bn2(self.conv2(x)))

        # Flatten the tensor
        x = x.view(x.size(0), -1)

        # Apply dropout to prevent overfitting
        x = self.dropout(x)

        # Pass through fully connected layers + layer norm + relu
        x = torch.relu(self.ln(self.fc1(x)))
        x = self.fc2(x)
        return x
```

```python
import torch
import torch.nn as nn

def calculate_conv_output_size(input_size, kernel_size, padding, stride=1):
    return (input_size + 2 * padding - kernel_size) // stride + 1

class CoolantTempCNN(nn.Module):
    def __init__(self, input_sequence_length, kernel_size1=3, kernel_size2=15):
        super(CoolantTempCNN, self).__init__()
        # Compute padding for conv1 based on kernel_size1
        padding1 = (kernel_size1 - 1) // 2
        # Local feature extraction with a small kernel size
        self.conv1 = nn.Conv1d(in_channels=4, out_channels=16, kernel_size=kernel_size1, stride=1, padding=padding1)
        self.bn1 = nn.BatchNorm1d(16)  # Batch normalization after conv1

        # Compute padding for conv2 based on kernel_size2
        padding2 = (kernel_size2 - 1) // 2
        # Global feature extraction with a larger kernel size
        self.conv2 = nn.Conv1d(in_channels=4, out_channels=16, kernel_size=kernel_size2, stride=1, padding=padding2)
        self.bn2 = nn.BatchNorm1d(16)  # Batch normalization after conv2

        # Dropout layer to prevent overfitting
        self.dropout = nn.Dropout(p=0.2)

        # Calculate the sequence length after each convolution
        sequence_length_after_conv1 = calculate_conv_output_size(
            input_sequence_length, kernel_size=kernel_size1, padding=padding1)
        sequence_length_after_conv2 = calculate_conv_output_size(
            input_sequence_length, kernel_size=kernel_size2, padding=padding2)

        # Fully connected layers (adjust input size based on conv1 and conv2 outputs)
        self.fc1 = nn.Linear(16 * (sequence_length_after_conv1 + sequence_length_after_conv2), 64)
        self.ln = nn.LayerNorm(64)  # Layer normalization before final layer
        self.fc2 = nn.Linear(64, 1)  # Regression output for predicting coolant temperature

    def forward(self, x):
        # Apply first convolutional layer (local features) + batch norm + relu
        x_local = torch.relu(self.bn1(self.conv1(x)))

        # Apply second convolutional layer (global features) + batch norm + relu
        x_global = torch.relu(self.bn2(self.conv2(x)))

        # Concatenate the outputs of both convolutions along the channel dimension
        x_concat = torch.cat((x_local, x_global), dim=2)

        # Flatten the tensor
        x_concat = x_concat.view(x_concat.size(0), -1)

        # Apply dropout to prevent overfitting
        x_concat = self.dropout(x_concat)

        # Pass through fully connected layers + layer norm + relu
        x_concat = torch.relu(self.ln(self.fc1(x_concat)))
        x_concat = self.fc2(x_concat)
        return x_concat
```

```python

# Sample data preparation
# df = pd.read_csv('your_data.csv')  # Load your CSV file with actual data

# Prepare the input (features) and output (target)
# X = df[['Vehicle_Speed', 'Current_gear_shift_position_(Current_gear)', 'time', 'initial_coolant_temp' , 'speed_diff' ,'gear_change']].values
X = df[['time', 'initial_coolant_temp' , 'speed_diff' ,'gear_change']].values
y = df['Coolant_temperature'].values  # Assuming you have this target column

# Convert the data to PyTorch tensors
X_tensor = torch.tensor(X, dtype=torch.float32).unsqueeze(2)  # Add extra dimension for CNN: [batch, channels, seq_len]
y_tensor = torch.tensor(y, dtype=torch.float32).unsqueeze(1)  # Target as a tensor

# Create dataset and data loader
dataset = TensorDataset(X_tensor, y_tensor)
dataloader = DataLoader(dataset, batch_size=64, shuffle=True)

# Get the input sequence length from the tensor (i.e., the sequence length for each sample)
input_sequence_length = X_tensor.shape[2]


# Initialize the model, loss function, and optimizer
# model = CoolantTempCNN(input_sequence_length)
model = CoolantTempCNN(input_sequence_length=input_sequence_length, kernel_size1=3, kernel_size2=15)

optimizer = optim.Adam(model.parameters(), lr=0.001)
optimizer = optim.Adam(model.parameters(), lr=0.01, weight_decay=1e-6)

criterion = nn.MSELoss()

# Training loop
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Set up the learning rate scheduler (decrease LR by a factor of 0.1 every 10 epochs)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=6, gamma=0.5)

# Training loop
num_epochs = 30  # Adjust as needed
for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    for inputs, targets in dataloader:
        inputs, targets = inputs.to(device), targets.to(device)

        # Forward pass
        outputs = model(inputs)
        loss = criterion(outputs, targets)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

    # Step the scheduler after each epoch
    scheduler.step()

    # Print the current learning rate
    current_lr = scheduler.get_last_lr()[0]
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss/len(dataloader)}, LR: {current_lr}")

# Save the trained model
model_filename = 'coolant_temp_cnn_model.pth'
torch.save(model.state_dict(), model_filename)
print(f"Model saved as {model_filename}")

```
